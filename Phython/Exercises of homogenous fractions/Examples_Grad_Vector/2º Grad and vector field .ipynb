## 1ยบ I.C

```python
import matplotlib.pyplot as plt
from scipy import *
from scipy import integrate
from scipy.integrate import ode
import numpy as np
import math

fig = plt.figure(num=1)
ax=fig.add_subplot(111)


## Vector field function
## x[0] = x // x[1] = y

def vf(t,x):
  dx=np.zeros(2)

  dx[0]=1
  dx[1]= np.sqrt((x[0]+x[1])/(2*x[0]))   ## <- Here equation of dy/dx
  return dx

##Solution curves
t0=0; tEnd=10; dt=0.01;
r = ode(vf).set_integrator('vode', method='bdf',max_step=dt)

##Initial conditions for ecah line with its color
ic=[[1,1]]
color=['r']
listY = [] ## Lista vacia para aรฑadir valores de Y del for..


for k in range(len(ic)):
    Y=[];T=[];S=[];
    r.set_initial_value(ic[k], t0).set_f_params()
    while r.successful() and r.t +dt < tEnd:
        r.integrate(r.t+dt)
        Y.append(r.y)

    S=np.array(np.real(Y))
    ax.plot(S[:,0],S[:,1], color = color[k], lw = 5.25)
    
    for elem in S[:,1]:
        listY.append(elem)
    

##Vector field with its linespace(interval)
X,Y = np.meshgrid( np.linspace(1,10,20),np.linspace(-5,10,20) )

V = np.sqrt((X+Y)/(2*X))  ## <- Here equation of dy/dx


##Normalize arrows not to have long arrows similar
U = 1
N = np.sqrt(U**2+V**2)  
U2, V2 = U/N, V/N
ax.quiver( X,Y,U2, V2)

##Analytical solution
xs =  S[:,0]
y_exact = xs
plt.plot(xs, y_exact, "-", color = "b")
y_exact2 = -xs/2
plt.plot(xs, y_exact2, "-", color = "g");


##Limits of the graph
plt.xlim([1,10])
plt.ylim([-5,10])
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
plt.show()
```

```python
y_diff = np.abs(y_exact-listY)
plt.semilogy(xs, y_diff)
plt.ylabel("Error")
plt.xlabel("x")
plt.title("Error in numerical integration with first y_exact")
```

```python
y_diff = np.abs(y_exact2-listY)
plt.semilogy(xs, y_diff)
plt.ylabel("Error")
plt.xlabel("x")
plt.title("Error in numerical integration with second y_exact")
```

## 2ยบ I.C

```python
fig = plt.figure(num=1)
ax=fig.add_subplot(111)


## Vector field function
## x[0] = x // x[1] = y

def vf(t,x):
  dx=np.zeros(2)

  dx[0]=1
  dx[1]= np.sqrt((x[0]+x[1])/(2*x[0]))   ## <- Here equation of dy/dx
  return dx

##Solution curves
t0=0; tEnd=10; dt=0.01;
r = ode(vf).set_integrator('vode', method='bdf',max_step=dt)

##Initial conditions for ecah line with its color
ic=[[1,3]]
color=['r']


for k in range(len(ic)):
    Y=[];T=[];S=[];
    r.set_initial_value(ic[k], t0).set_f_params()
    while r.successful() and r.t +dt < tEnd:
        r.integrate(r.t+dt)
        Y.append(r.y)

    S=np.array(np.real(Y))
    ax.plot(S[:,0],S[:,1], color = color[k], lw = 1.25)
    

##Vector field with its linespace(interval)
X,Y = np.meshgrid( np.linspace(0,2,20),np.linspace(-1,10,20) )

V = np.sqrt((X+Y)/(2*X))  ## <- Here equation of dy/dx


##Normalize arrows not to have long arrows similar
U = 1
N = np.sqrt(U**2+V**2)  
U2, V2 = U/N, V/N
ax.quiver( X,Y,U2, V2)


##Limits of the graph
plt.xlim([0,2])
plt.ylim([-1,10])
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
plt.show()
```

```python
## k levels?
levels_solution = []
xs,ys=1,3
K1= (((np.sqrt((xs+ys)/(2*xs) -1))**2) * np.abs((np.sqrt((xs+ys)/(2*xs)) +(1/2))))/(np.abs(xs)**(-3/2))
print('k1 value is',K1) 
levels_solution.append(K1)
```

```python
## Levels we are interested to compare solution from gradient field
print(levels_solution)
levels=sorted(levels_solution)
print(levels) ## Ordenado para plotear las contour curves
```

```python
%matplotlib inline
plt.style.use('seaborn-white')

xs = np.linspace(0, 2, 20)
ys = np.linspace(-1,10, 20)

def fs(xs, ys):  ## y_exact equation without constant
    
    return (((np.sqrt((xs+ys)/(2*xs) -1))**2) * np.abs((np.sqrt((xs+ys)/(2*xs)) +(1/2))))/(np.abs(xs)**(-3/2))

X, Y = np.meshgrid(xs, ys)
Z = fs(X,Y)

contours = plt.contour(X, Y, Z, levels, colors='blue') ## Our solution
plt.clabel(contours, inline=1, fontsize=10)

plt.contourf(X, Y, Z, 20, cmap='RdGy') ## General contour curve
plt.colorbar()
```
