```python
import matplotlib.pyplot as plt
from scipy import *
from scipy import integrate
from scipy.integrate import ode
import numpy as np

fig = plt.figure(num=1)
ax=fig.add_subplot(111)


## Vector field function
## x[0] = x // x[1] = y

def vf(t,x):
  dx=np.zeros(2)

  dx[0]=1
  dx[1]= (x[0]+x[1])/(x[0]-x[1])   ## <- Here equation of dy/dx
  return dx

##Solution curves
t0=0; tEnd=10; dt=0.01;
r = ode(vf).set_integrator('vode', method='bdf',max_step=dt)

##Initial conditions for ecah line with its color
ic=[[2,0], [-5,0], [-5,-2.5],[2,-2]]
color=['r','b','g','k']


for k in range(len(ic)):
    Y=[];T=[];S=[];
    r.set_initial_value(ic[k], t0).set_f_params()
    while r.successful() and r.t +dt < tEnd:
        r.integrate(r.t+dt)
        Y.append(r.y)

    S=np.array(np.real(Y))
    ax.plot(S[:,0],S[:,1], color = color[k], lw = 1.25)
    

##Vector field with its linespace(interval)
X,Y = np.meshgrid( np.linspace(-5,5,20),np.linspace(-5,5,20) )

V = (X+Y)/(X-Y)  ## <- Here equation of dy/dx


##Normalize arrows not to have long arrows similar
U = 1
N = np.sqrt(U**2+V**2)  
U2, V2 = U/N, V/N
ax.quiver( X,Y,U2, V2)

#Anals solutions


##Limits of the graph
plt.xlim([-5,5])
plt.ylim([-5,5])
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
plt.show()
```

```python
## k levels?7
levels_solution = []
x,y=2,0
K1= -np.log(np.abs(np.sqrt(x**2 + y**2))) +np.arctan(y/x)
print('k1 value is',K1) 
levels_solution.append(K1)

x,y=-5,0
K2= -np.log(np.abs(np.sqrt(x**2 + y**2))) +np.arctan(y/x)
print('k2 value is',K2)
levels_solution.append(K2)

x,y=-5,2.5
K3= -np.log(np.abs(np.sqrt(x**2 + y**2))) +np.arctan(y/x)
print('k3 value is',K3)
levels_solution.append(K3)

x,y=2,-2
K4= -np.log(np.abs(np.sqrt(x**2 + y**2))) +np.arctan(y/x)
print('k4 value is',K4)
levels_solution.append(K4)
```

```python
## Levels we are interested to compare solution from gradient field
print(levels_solution)
levels=sorted(levels_solution)
print(levels) ## Ordenado para plotear las contour curves
```

```python
%matplotlib inline
plt.style.use('seaborn-white')

xs = np.linspace(-5, 5, 20)
ys = np.linspace(-5, 5, 20)

def fs(xs, ys):  ## y_exact equation without constant
    
    return -np.log(np.abs(np.sqrt(xs**2 + ys**2))) +np.arctan(ys/xs)

X, Y = np.meshgrid(xs, ys)
Z = fs(X,Y)

contours = plt.contour(X, Y, Z, levels, colors='blue')
plt.clabel(contours, inline=1, fontsize=10)

plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar()
```
